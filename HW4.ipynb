```python
# retreived from https://www.ncdc.noaa.gov/cdo-web/search
df = pd.read_csv("BOI_WEATHER.csv")
df.head()
```

```python
df.drop('STATION', inplace=True, axis=1)
df.drop('NAME', inplace=True, axis=1)
df.head()
```

```python
df["RAIN_BOOL"] = ""
df.head()
```

```python
for index in df.index:
    if df.loc[index, 'PRCP'] > 0.00:
       df.loc[index, 'RAIN_BOOL'] = 1
    else: df.loc[index, 'RAIN_BOOL'] = 0
```

```python
df[['DATE', 'PRCP', 'RAIN_BOOL']].head()
```

```python
# A lot of weird data points in the dataset, that I don't think really matter.
# There is one called 'weather type' that I would imagine says things like
# 'cloudy', 'sunny', 'partly cloudy' etc, but I don't want to deal with 
# one hot encoding right now.
features = ['AWND', 'PRCP', 'SNOW', 'TMAX', 'TMIN']
label = ['RAIN_BOOL']

X = df[features].values
Y = df[label].values
Y=Y.astype('int')

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3)
```

```python
model = LogisticRegression(max_iter=5000)
model.fit(X_train, Y_train.ravel())
```

```python
# Yeah not too bad!
# I didn't really want to deal with cleaning data so I took out some 
# features with NAN values etc., so that could be contributing to the error.
# Overall though i'm pleased, and I'd say that 
print("Accuracy:",metrics.accuracy_score(Y_test, model.predict(X_test)))
```
